<h2>Contexto
<p><h4>De acuerdo con la Organización Mundial de la Salud, en el mundo viven cerca de 700 millones de adultos entre 30 y 79 años que sufren de hipertensión. 
    La hipertensión es una enfermedad no transmisible con causas  multifactoriales, la cual puede ser controlable con medicamentos y un estilo de vida saludable.
    Sin embargo, es una enfermedad que rara vez presenta síntomas (asintomática), por lo que su detección podría darse después de presentar algunas complicaciones.
    La hipertensión es una de las principales causas de muerte en el mundo y su padecimiento aumenta el riesgo de padecer enfermedades cardiovasculares como infartos, accidentes cerebrovasculares, enfermedad renal crónica, encefalopatías, etc. 
    Es por ello que es importante tanto prevenir su aparición como diagnosticarla a tiempo.
