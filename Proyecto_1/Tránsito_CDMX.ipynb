<h2>Contexto
    <h4>Día a día, ocurren accidentes de tránsito en cualquier parte del país. <h2>Tipo de Evento

<h4>En la siguiente gráfica, se agruparán los tipos de evento que hubo por año, de tal forma que sea más fácil visualizar qué eventos fueron los más comunes a lo largo de los años. 

```python
df3=df2.copy()
```

```python

df3['año']=df3['fecha_evento'].dt.year
```

```python
#creamos una función que me va a devolver una serie (que es un array unidimensional, o sea, una tablita con índices y una columna)
def grupo(x):
    nombres={
        'total_evento':x['tipo_de_evento'].size

 }
    return pd.Series(nombres, index=['total_evento']) #lo que quiero es que devuelva mis valores de mi diccionario en una serie que luego se puede
    #añadir a mi dataframe. No olvidar que primero señalo los datos y luego el índice (que en este caso es el nombre de la columna)
```

```python
#Invocamos a la función
datos2=df3.groupby(['año','tipo_de_evento']).apply(grupo)
print(datos2)
```

```python
sns.lineplot(data=datos2, x="año", y="total_evento", hue="tipo_de_evento")
```

<h4>Como se puede apreciar en el gráfico, CHOQUE es el evento más común. Si bien ATROPELLADO se mantuvo en segundo lugar, a partir del 2020 fue superado por DERRAPE. 
Algo interesante a notar es que a partir de la pandemia el número de choques aumentaron drásticamente (algo a considerar es que el tráfico en la CDMX aumentó en ese mismo año)

<h2>Días con más accidentes

```python
#aquí primero tuve que importar datetime 
dia_prueba=df2['fecha_evento']
```

```python
dias=dia_prueba.apply(lambda x: x.strftime("%A"))
#aquí lo que hice fue aplicar el método strftime a la variable con mi fecha, que es dia_prueba, el cual devuelve el día de la semana y se pone el %A para ello. 
#en relacion al apply, es aplicar una función a la columna
```

```python
dia_semana=dias.replace({"Saturday":'Sábado', 'Sunday':'Domingo','Friday':'Viernes','Monday':'Lunes','Tuesday':'Martes','Wednesday':'Miércoles','Thursday':'Jueves'}).value_counts()#con value_counts ya cuento mis días de la semana
```

```python
dia_semana
```

```python
dia_semana.plot(kind='bar')
```

<h4>En el gráfico se puede apreciar que en los días "SÁBADO" se registraron más tipo de accidente; en cambio, el día donde hubo menos accidentes fue "MARTES"
